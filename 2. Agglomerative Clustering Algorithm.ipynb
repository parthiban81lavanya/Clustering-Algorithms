```python
#Agglomerative Clustering Algorithm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3: 5].values # : row select all , 3rd and 4th column select only that column
```

```python
X
```

```python
from sklearn.cluster import AgglomerativeClustering

import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
clusmodel = AgglomerativeClustering(n_clusters = 5)
label = clusmodel.fit_predict(X)
```

```python
clusmodel
```

```python
unsupervised = pd.DataFrame(dataset)
```

```python
unsupervised
```

```python
unsupervised['Cluster_group']=label
```

```python
label
```

```python
unsupervised.to_csv("cluster.csv",index=False)
```
